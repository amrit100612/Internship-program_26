```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Histogram
plt.hist(data,bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()
```

```python
#Bar Plot
categories = ['A', 'B', 'C']
category_counts = [300,400,300]
plt.bar(categories, category_counts)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Plot of Categories')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns

iris = datasets.load_iris()
species = iris.target_names
sepal_count = np.bincount(iris.target)

plt.figure(figsize=(8, 8))
plt.pie(sepal_count, labels=species, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Iris Species')
plt.show()
```

```python
iris = datasets.load_iris()
species = iris.target_names
species_counts = np.bincount(iris.target)

#Donut charts
plt.figure(figsize = (10,8))
plt.pie(species_counts, labels = species, autopct = '%1.1f%%', startangle = 90, wedgeprops = {'width': 0.5})
plt.title('Species Distribution in Iris Dataset(Donut Chart)')
plt.show()
```

```python
#Pie Diagram

color = ['#ff8888', '#ffc010','#1110FF']
explode = (0,0.1,0.1)

plt.pie(category_counts, explode= explode, labels = categories, colors=color, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.title('fruit Distribution')
plt.show()
```

```python
#Bivariate cases

import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
plt.scatter(iris.data[:,0],iris.data[:,2], c = iris.target)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.show()
```

```python
iris = datasets.load_iris()

df1 = pd.DataFrame(iris.data, columns = iris.feature_names)
df1['species'] = iris.target_names[iris.target]

plt.figure(figsize= (8,6))
for species in iris.target_names:
   sns.kdeplot(data = df1[df1['species'] == species], x = 'sepal length (cm)', label = species, fill = True)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.title('Distribution of Sepal Length by Species')
plt.legend()
plt.show()
```

```python
X = ['Section A','Section B','Section C','Section D']
Ygirls = [10,20,30,40]
Zboys = [20,30,20,30]
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls',color = 'r')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Student")
plt.title("Number of Student in each group")
plt.legend()
plt.show()
```

```python
data = {
    'city' : ['City A','City B','City C','City D','City E'],
    'population' : [25,28,30,40,50],
    'Avg_Income' : [40000,50000,60000,55000,70000],
    'Pollution_Index' : [55,66,80,75,35]
}
df = pd.DataFrame(data)
df

plt.figure(figsize = (3,3))
plt.scatter(df['Avg_Income'],df['Pollution_Index'],s = df['population']* 0.5, alpha = 0.5, c = df['population'],cmap="viridis")
plt.colorbar(label = 'Population')
plt.xlabel('Average Income (In $1000)')
plt.ylabel('Pollution Index')
plt.title('City Bubble PLot : Income vs Pollution')

for i in range(len(df)):
  plt.text(df['Avg_Income'][i], df['Pollution_Index'][i], df['city'][i], fontsize= 9,ha ='right')

plt.show()
```
