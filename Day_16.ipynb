Comparative Look at Cloud Deployment :
======================================

* Public Cloud: Service are accessible to the general public

* Private Cloud: Service are accessible only within single org.

* Hybrid Cloud: Micture of public and private cloud

* Community Cloud: Service are accessible to a specific group of org.with shared concerns



Cloud Service Stack :   
==============================

* Saas(Software as a service)
* Paas (Platform as a Service)
* Iaas (Infrastructure as a service)

Iaas :   
======  
Rent the hardware but manage the operating sysytem and applications

What you Get:
* VM & Physical Servers
* Virtual Storage & Networks(VLANs)
* Load Balancers, IP addresses

Paas :   
======  
Complete environmentto support the lifecycle of building web applications,without buying and managing the underlying infrastructure

What you Get:
* Runtime environment for applications
* Deployment and Deployment tools
* built in APIs for creating web-based applications
