```python
import numpy as np
a = np.array([[1,2],[7,8]])
b=  np.array([[4,5],[9,1]])
c= np.matrix([[1,2],[3,4]])
d= np.matrix([[5,6],[8,9]])
print("\nC", type(c))
print(c)
print("\nD", type(d))
print(d)
print("\n Operation on two ndarray obj. (Elementwise)")
print(a*b)
print("\n Operation on two matrix obj. (same as np.dot())")
print(c*d)
```

```python
import numpy as np
a = np.array([[1,2],[3,4],[5,6]])
print("a= ")
print(a)

print("\n With np Transpose function = ")
print(np.transpose(a))

print("\n With Ordinary Transpose method = ")
print(a.transpose())

print("\n With ndarray.1 Short form = ")
print(a.T)
```

```python
#rank
import numpy as np
a = np.arange(1,10)
a.shape = (3,3)
print("a = ")
print(a)
print("rank of a : ")
print(np.linalg.matrix_rank(a))
```

```python
#Determinant

import numpy as np
a=np.array([[2,2,1],[1,3,1],[1,2,2]])
det=np.linalg.det(a)
print(f"Determinant of a matrix:\n{det}")


```

```python
#Inverse
import numpy as np
a=np.array([[2,2,1],[1,3,1],[1,2,2]])
inv=np.linalg.inv(a)
print(f"Inverse of a matrix:\n{inv}")
```

```python
#Det=0 Error
import numpy as np
a=np.array([[2,8],[1,4]])
inv=np.linalg.inv(a)
print(f"Inverse of a matrix:\n{inv}")
```

```python
#psudo Inverse
import numpy as np
a=np.array([[2,8],[1,4]])
pinv=np.linalg.pinv(a)
print(f"Inverse of a matrix:\n{pinv}")
```

```python
#flattening
import numpy as np
a = np.arange(1,10)
a.shape = (3,3)
print("a = ")
print(a)
print("\nAfter flattening: ")
print(a.flatten())
print("\nAfter reshaping:")
print(a.reshape(-1))
```

```python
#Eigen value
import numpy as np
a=np.array([[2,2,1],[1,3,1],[1,2,2]])
w,v = np.linalg.eig(a)
print("Eigen Value")
print(w)
print("Eigen Vector")
print(v)
```
