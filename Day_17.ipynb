Building a Digital Fortress:
===================================
1. Defence in depth
2. least Privilege
3. Separation of Duties
4. Security Key Design
5. Simplicity(KISS: keep it simple and stupid)


* Principal 1: Defence in Depth- Assume at any single point and build layers

* Principal 2: The Principle of Least Privilege - Limit the blast radius of a compromise by strictly restricting access

* Principal 3: Separation of Duties - Eliminate single point of failure through checks and balances

* Principal 4: Security by Design - Integration beats retrofitting: build security in from the start

* Principal 5: KISS - Complexity is the enemy of security


Conclusion:
=======================

* Multiple layer - No single control is foolproof
* Minimun access - reduce potential damage from compromise
* Divided Responsibilities - Prevent Single point of failure
* Built in Design - Retrofitting is costly and less effective
* Simplicity - Complexity breeds vulnerability
