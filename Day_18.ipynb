The ENGINE : How Virtualization Works
=================

* Definition : Process of taking one physical computer and making act like multiple separate computers

* The Hypervisor: Special software (like VMalware ESXI or KVM ) that acts as a "pizza cutter " slicing the physical server into isolated Virtual  Machine(VMS)

* Component 1: Compute (Brain)
=========
General Purpose Compute -
* Use Case: Everyday business apps, web servers,email
* Spaces : Balanced CPU, memory, and storage

  High-Performance Computing -
* Use Case : Weather forecasting , DNA sequencing, AI training
* Spaces : High clock speed,massive core counts, immense power

* Component 2: Storage (Where data lives)
=========
Storage Unit : Cheap,low performance,best for things you dont access constantly,backups,image,videos,docs

  Hard Drive: Fast,expensive,high performance,best for databases and OS installation requiring speed

  Shared Folder : Shared access,best for team collaboration and content management system.

* Component 3: Networking
=
  Controlling Virtual infrastructure and controlling data flow

    FLOWCHART : Internet -> Firewall -> Load balancer -> VM's -> Storage
