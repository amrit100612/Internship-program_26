```python
avg_item_visibility = data.groupby(['Item_Identifier'])['Item_Visibility'].mean().reset_index()
avg_item_visibility
```

```python
def get_item_visibility(x):
  return avg_item_visibility.loc[avg_item_visibility['Item_Identifier']==x,'Item_Visibility'].values[0]
```

```python
data['avg_item_visibility']=data['Item_Identifier'].apply(get_item_visibility)
data.head()
```
