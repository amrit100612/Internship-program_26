```python
def median(l):

  l=[26,3,5,6,7,8]
  n= len(l)
  x=sorted(l)
  if n%2==1:
    return x[n // 2]
  else:
    mid1=x[n//2]
    mid2=x[n//2-1]
    return(mid1+mid2)/2
print("The Median is :",median(l))

```

```python
l = [26,26,3,5,6,7,8]
counts = {}
for x in l:
   counts[x] = counts.get(x,0)+1
mode = max(counts)
print("The Mode is : ", mode)

```

```python
l = [26,26,3,5,6,7,8]
k=sum(l)
n= len(l)
mean=k/n
var = sum((mean-x)**2 for x in l) / n-1
print(var)

SD= (var**(1/2))
print(SD)

cov=(SD*100)/mean
print (cov)
```

```python
import random
import math


random.seed(52)
heights = [random.randint(150, 190) for i in range(20)]
print(heights)

mean = sum(heights) / len(heights)
print(mean)

vari = sum([(height - mean) ** 2 for height in heights]) / (len(heights) - 1)
print(vari)

sd  = math.sqrt(vari)
print(sd )
```
