### Dataset Overview and Structural Summary

The dataset consists of **614 observations (rows)** and **13 variables (columns)**, representing applicant-level information for a loan approval prediction problem. Each row corresponds to an individual loan application, and the dataset includes demographic, financial, and credit-related attributes along with the loan approval status.

The total memory usage of the dataset is approximately **62.5 KB**, indicating that it is a relatively small, structured tabular dataset suitable for exploratory data analysis (EDA) and machine learning modeling without high computational overhead.

---

### Variable Classification and Data Types

The dataset contains a mix of **categorical and numerical features**, structured as follows:

#### 1. Categorical Variables (8 columns – object type)

* **Loan_ID** – Unique identifier for each loan applicant (non-null: 614)
* **Gender** – Applicant’s gender (601 non-null)
* **Married** – Marital status (611 non-null)
* **Dependents** – Number of dependents (599 non-null)
* **Education** – Education level (614 non-null)
* **Self_Employed** – Self-employment status (582 non-null)
* **Property_Area** – Property location category (614 non-null)
* **Loan_Status** – Target variable indicating loan approval (614 non-null)

These variables represent qualitative characteristics of applicants and play a critical role in classification-based predictive modeling.

---

#### 2. Numerical Variables (5 columns)

**Integer Variable (1 column):**

* **ApplicantIncome** – Applicant’s income (614 non-null)

**Floating-Point Variables (4 columns):**

* **CoapplicantIncome** – Co-applicant’s income (614 non-null)
* **LoanAmount** – Loan amount requested (592 non-null)
* **Loan_Amount_Term** – Loan repayment term (600 non-null)
* **Credit_History** – Credit history indicator (564 non-null)

These numerical attributes represent financial capacity and creditworthiness, which are significant predictors in loan approval models.

---

### Missing Data Analysis

Although the dataset is largely complete, several variables contain missing values:

* **Credit_History** – 50 missing values (most significant)
* **Self_Employed** – 32 missing values
* **LoanAmount** – 22 missing values
* **Dependents** – 15 missing values
* **Loan_Amount_Term** – 14 missing values
* **Gender** – 13 missing values
* **Married** – 3 missing values

The presence of missing data, particularly in **Credit_History**, is critical because credit history is typically one of the most influential factors in loan approval decisions. Proper imputation strategies (mean/median/mode imputation or predictive imputation) will be necessary before model training.

---

### Target Variable Identification

The **Loan_Status** column serves as the dependent (target) variable in this dataset. It is categorical and represents whether a loan application was approved or rejected. Since the target is categorical, this dataset is appropriate for **classification-based machine learning algorithms**, such as:

* Logistic Regression
* Decision Tree
* Random Forest
* Support Vector Machine
* Gradient Boosting Models

Thus, the dataset integrates both **behavioral and financial attributes**, making it suitable for predictive modeling in credit risk assessment.

# Report:

This code handles missing values in the Loan Prediction dataset.

For categorical features such as Gender, Married, Dependents, Self_Employed, and also for Loan_Amount_Term and Credit_History, missing values are filled using the mode (most frequent value).

Mode is chosen because it preserves the most common category and avoids data distortion.

For the numerical feature LoanAmount, missing values are replaced using the median, which is robust to outliers.

This step ensures there are no missing values in important columns, making the dataset clean and ready for machine learning models.

# Trained the model for Machine Learning

```python
#imbalance class

loan['Loan_Status'].value_counts()
```

```python
X = loan_encoded.drop('Loan_Status', axis=1)
y = loan_encoded['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE


smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

y_resampled.value_counts()
```

```python
X
```


---

The target variable **Loan_Status** showed class imbalance, where 422 loans were approved and 192 were not approved. This imbalance can cause the model to become biased toward the majority class, resulting in misleading accuracy and poor prediction of rejected loans. To address this issue, **SMOTE (Synthetic Minority Oversampling Technique)** was applied. SMOTE generates synthetic samples for the minority class instead of simply duplicating data, helping to balance the dataset effectively. After applying SMOTE, both classes contained equal samples, improving model fairness, recall, and overall predictive performance.


```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled,test_size=0.2, random_state=42)
```

```python
#import all requirements

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report,confusion_matrix


# import all models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
#work without hypermetertunning athen check 1 by one model

#work on class imbalance
#"""smote = SMOTE(random_state=42)"""search this"""
```

```python
#solver=['liblinear', 'newton-cg', 'lbfgs', 'sag', 'saga']
model = LogisticRegression(solver = 'liblinear')
model.fit(X_train, y_train)
```

```python
log_predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, log_predictions)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, log_predictions))
```

```python
model = LogisticRegression(max_iter=100)
model.fit(X_train, y_train)
log_predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, log_predictions)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, log_predictions))
```

```python
#Decision Tree

df_model = DecisionTreeClassifier()
df_model.fit(X_train, y_train)
```

```python
df_predicition = df_model.predict(X_test)
accuracy = accuracy_score(y_test, df_predicition)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, df_predicition))
```

```python
#Random Forest

rf_model = RandomForestClassifier(n_estimators=500)
rf_model.fit(X_train, y_train)
```

```python
rf_predicition = rf_model.predict(X_test)
accuracy = accuracy_score(y_test, rf_predicition)
print(f"Accuacy: {accuracy}")
print(classification_report(y_test, rf_predicition))
```

```python
#support vector machine

svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
svm_predicition = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, svm_predicition)
print(f"Accuracy : {accuracy}")
print(classification_report(y_test, svm_predicition))
```

```python
def evaluate_model(y_true, y_pred):

  return{
    'Accuracy' : accuracy_score(y_true, y_pred),
    'Precision' : precision_score(y_true, y_pred, average='weighted'),
    'Recall' : recall_score(y_true, y_pred, average='weighted'),
    'F1score' : f1_score(y_true, y_pred, average='weighted')
  }

results = {
    'Logistic Regression' : evaluate_model(y_test, log_predictions),
    'Decision Tree' : evaluate_model(y_test, df_predicition),
    'Random Forest' : evaluate_model(y_test, rf_predicition),
    'Support Vector Machine' : evaluate_model(y_test, svm_predicition)
}

results_df = pd.DataFrame(results).T
results_df
```
