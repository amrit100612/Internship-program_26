```python
#make a copy of the  data
clean_data=df1.copy()
```

```python
new_col_name=[col.lower() for col in clean_data.columns]
clean_data.columns=new_col_name
print(clean_data.columns)
```

```python
#fill Null Value by taking mean
clean_data['item_weight'].fillna(clean_data['item_weight'].mean(),inplace=True)
clean_data.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.kdeplot(clean_data['item_weight'], fill=True)
plt.title('KDE Plot of Item Outlet Sales')
plt.xlabel('Item Outlet Sales')
plt.ylabel('Density')
plt.show()
```

```python
outlet_size_mode = clean_data.pivot_table(values='outlet_size', columns=['outlet_type'], aggfunc=(lambda x: x.mode()))
outlet_size_mode
```

```python
missing_value = clean_data['outlet_size'].isnull()
clean_data.loc[missing_value,'outlet_size'] = clean_data.loc[missing_value,'outlet_type'].apply(lambda x: outlet_size_mode[x].outlet_size)
clean_data.isna().sum()
```

```python
pd.pivot_table(df1,index=['Outlet_Size','Outlet_Type'],aggfunc='size')
```

```python
#item_visibility

print('Total of 0s before replace:',sum(clean_data['item_visibility']==0))

#replce 0s with the mean
clean_data['item_visibility'] = clean_data['item_visibility'].replace(to_replace=0 , value = clean_data['item_visibility'].mean())

#print total number of 0s after replace
print('Total of 0s after replace:',sum(clean_data['item_visibility']==0))
```
